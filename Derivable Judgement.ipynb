```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 120

data = {
    "record_id": [f"REC{i+1:03d}" for i in range(n)],
    "age": np.random.randint(18, 70, n),
    "weight": np.random.randint(45, 100, n),
    "gender": np.random.choice(["Male", "Female"], n),
    "region": np.random.choice(["North", "South", "East", "West"], n),
    "smoking_status": np.random.choice(
        ["Smoker", "Non-Smoker", "Former Smoker"], n, p=[0.35, 0.45, 0.20]
    ),
    "exercise_frequency": np.random.choice(
        ["Daily", "Weekly", "Rarely", "Never"], n, p=[0.3, 0.35, 0.2, 0.15]
    ),
}

df = pd.DataFrame(data)

# Derived fields
df["bmi"] = np.round(df["weight"] / ((np.random.uniform(1.5, 1.8, n)) ** 2), 1)
df["blood_pressure"] = np.random.normal(130, 15, n)
df["cholesterol_level"] = np.random.normal(210, 30, n)
df["glucose_level"] = np.random.normal(110, 25, n)

# Diseases (probability-based)
df["diabetes"] = np.where((df["glucose_level"] > 126) | (df["bmi"] > 30), True, False)
df["hypertension"] = np.where(df["blood_pressure"] > 140, True, False)

# Age groups
df["age_group"] = pd.cut(
    df["age"],
    bins=[17, 25, 35, 45, 60, 100],
    labels=["18-25", "26-35", "36-45", "46-60", "60+"]
)

# Dates
df["visit_date"] = pd.date_range(start="2023-01-01", periods=n, freq="D")

df.to_csv("health_records.csv", index=False)

df.head()
```

Part B-Data Analysis & Testing Tasks:

1. Example 1 – Exercise vs Hypertension

Ho: Exercise frequency and hypertension status are independent.

H1: Exercise frequency and hypertension status are associated.

(Chi-square)

2. Example 2 – BMI Difference Between Genders

H₀: There is no significant difference in mean BMI between genders.

H₁: There is a significant difference in mean BMI between genders.

(T-test)

```python
#Calculate Confidence Interval 

from scipy import stats

confidence=0.95

#age
mean_of_age=df['age'].mean()
sem_of_age=stats.sem(df['age'])
ci_of_age = stats.t.interval(confidence,len(df)-1,mean_of_age,sem_of_age)

#weight
mean_of_weight=df['weight'].mean()
sem_of_weight=stats.sem(df['weight'])
ci_of_weight = stats.t.interval(confidence,len(df)-1,mean_of_weight,sem_of_weight)

ci_of_age, ci_of_weight
```

```python
#Find Critical Value and P-Value

critical_value=stats.t.ppf(1-0.05/2,len(df)-2)
critical_value
```

```python
#Perform Z test or T-test 

bmi_gender_male=df[df['gender']=='male']['bmi']
bmi_gender_female=df[df['gender']=='female']['bmi']

tstate, pvalue = stats.ttest_ind(bmi_gender_male,bmi_gender_female,equal_var=False)
#tstate, pvalue

if pvalue < 0.05:
    print("Reject H0")
    print("There is a significant difference in mean BMI between genders.")
else:
    print("Fail to Reject H0")
    print("There is no significant difference in mean BMI between genders.")
```

```python
#Chi-Square
contigency_table=pd.crosstab(df["exercise_frequency"],df["hypertension"])
contigency_table

chi_sq, p, dof, expected = stats.chi2_contingency(contigency_table)

chi_sq, p

if p < 0.05:
    print("Reject H0")
    print("Exercise frequency and hypertension status are associated.")
else:
    print("Fail to Reject H0")
    print("Exercise frequency and hypertension status are independent.")
```

```python
#Anova
bmi_low = df[df["exercise_frequency"] == "Never"]["bmi"]
bmi_medium = df[df["exercise_frequency"] == "Weekly"]["bmi"]
bmi_high = df[df["exercise_frequency"] == "Rarely"]["bmi"]


f_stat, p_value = stats.f_oneway(bmi_low, bmi_medium, bmi_high)

print("F-statistic:", f_stat)
print("p-value:", p_value)


if p_value < 0.05:
    print("Reject H0")
    print("There is a significant difference in mean BMI among exercise groups.")
else:
    print("Fail to Reject H0")
    print("No significant difference in mean BMI among exercise groups.")
```
